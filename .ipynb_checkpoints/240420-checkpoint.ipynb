# logical opers

```python
def logical_calc(array, op):
    if len(array) > 1:
        if op == "OR":
            curr = (array[0] | array[1])
            if len(array) > 2:
                for i in range(2, len(array)):
                    curr = curr | array[i]


        if op == "AND":
            curr = (array[0] & array[1])
            if len(array) > 2:
                for i in range(2, len(array)):
                    curr = curr & array[i]

        if op == "XOR":
            curr = (array[0] ^ array[1])
            if len(array) > 2:
                for i in range(2, len(array)):
                    curr = curr ^ array[i]
    else:
        return array[0]
    
    return curr
        
print(logical_calc([True, True, True, False, True, True], "XOR"))
```
